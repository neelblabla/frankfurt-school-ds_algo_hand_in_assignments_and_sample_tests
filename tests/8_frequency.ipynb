# Calculate the frequency of items in a given sequence
Given a sequence of elements (that support equality function), calculate the frequency of
each distinct element. For example, given the sequence [‘hi’, ‘I’, ‘am’, ‘Alexa’, ‘I’, ‘would’, ‘just’,
‘like’, ‘to’, ‘say’, ‘hi’]. The output should be:
hi 2
I 2
am 1
Alexa 1
would 1
just 1
like 1
to 1
say 1

```python
import pandas as pd
```

```python
a = ['hi', 'I', 'am', 'Alexa', 'I', 'would', 'just','like', 'to', 'say', 'hi']
b = [0]*len(a)
b
```

```python
df = pd.DataFrame(a, columns = ['Element'])
df['Count'] = b
df
```

```python
df = df.drop_duplicates()
df
```

```python
for index, row in df.iterrows():
    i = index
    #print(row['Element'])
    while (i < len(a)):
        if (df.Element[index] == a[i]):
            df.Count[index] = df.Count[index] + 1
        i = i + 1
df
```
